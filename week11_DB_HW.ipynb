```python

class Employee:
    """A sample Employee class"""

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay

    @property
    def email(self):
        return '{}.{}@email.com'.format(self.first, self.last)

    @property
    def fullname(self):
        return '{} {}'.format(self.first, self.last)

    def __repr__(self):
        return "Employee('{}', '{}', {})".format(self.first, self.last, self.pay)
```

# Exercise 1
Displaying the status of each student based on the credits they currently have

```python
conn = sqlite3 . connect ('AdvancedSQLLab.db')
# Execute INNER JOIN query and visualize the result
query = """
    SELECT ID, name,
        CASE
            WHEN credit < 30 then 'Freshman'
            WHEN credit < 60 then 'Sophomore'
            WHEN credit < 90 then 'Junior'
            WHEN credit < 120 then 'Senior'
            ELSE 'Graduate'
        END AS status
    FROM student
     """
df = pd . read_sql_query ( query , conn )
df
```

# Exercise 2
finding the number of students with each status

```python
conn = sqlite3 . connect ('AdvancedSQLLab.db')
# Execute INNER JOIN query and visualize the result
query = """
    SELECT
        CASE
            WHEN credit < 30 then 'Freshman'
            WHEN credit < 60 then 'Sophomore'
            WHEN credit < 90 then 'Junior'
            WHEN credit < 120 then 'Senior'
            ELSE 'Graduate'
        END AS status,
        COUNT(ID)
    FROM student
        GROUP BY status;
     """
df = pd . read_sql_query ( query , conn )
df
```
